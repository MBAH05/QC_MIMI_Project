```python
from qiskit.circuit import QuantumCircuit, QuantumRegister, AncillaRegister
from qiskit.quantum_info import Operator, Statevector

import numpy as np
```

```python
### Quantum circuit creation
```

```python
def multi_controlled_u_gate(n, U_matrix):
    """
    Function to create a multi-controlled U gate.

    Parameters:
        n (int): The number of control qubits.
        U_matrix (ndarray): The unitary matrix for the operation U.
    """
    # Create quantum circuit with n control qubits, 1 target qubit and ancillas
    circuit = QuantumCircuit(n + 1, n)  # n controls, 1 target

    # Apply the multi-controlled Toffoli gates
    for i in range(n - 1):
        circuit.ccx(i, i + 1, n)  # Apply Toffoli gates (3-qubit control)

    # Apply the unitary operation U on the target qubit when all control qubits are 1
    # Convert U matrix to an operator
    U = Operator(U_matrix)
    circuit.append(U, [n])  # Apply U to the target qubit

    # Draw the quantum circuit
    circuit.draw(output='mpl')

    return circuit
```

```python
### An examples to run and test the code
```

```python
# Define a simple unitary operation U (for example, a Pauli-X gate)
U_matrix = np.array([[0, 1], [1, 0]])  # Pauli-X gate

# Create a multi-controlled U gate with 3 control qubits and the Pauli-X gate
n = 3
circuit = multi_controlled_u_gate(n, U_matrix)

# Simulate and view the circuit
circuit.draw(output='mpl')
```
